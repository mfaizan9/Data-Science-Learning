```python
for i in df.columns:
    print(i,': ',df[i].nunique())

# Since number of unique values in PROVINCE & DIVISION is small than the rest of parameters,
# Hence we can group our data based on either PROVINCE or DIVISION    
```

```python
df.groupby(['DIVISION']).size().sort_values()
```

```python
# see the population division wise
df.groupby(['DIVISION'])['ALL SEXES (RURAL)'].mean()

## division ki base pe 'All sexes (rural)' ka mean
```

```python
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)

df.groupby(['PROVINCE' , 'DIVISION' , 'DISTRICT' , 'SUB DIVISION'])['ALL SEXES (RURAL)'].mean()

```

```python
# add values of two columns and store it in third
df['Total Population'] = df[['ALL SEXES (RURAL)', 'ALL SEXES (URBAN)']].sum(axis=1)
```

```python
# delete column in df
df.drop(['Total Urban'], axis = 1, inplace = True)
```
