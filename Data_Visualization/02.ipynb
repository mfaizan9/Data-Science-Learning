```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Change font to times for plots
plt.rcParams['font.family'] = 'Times New Roman'
# Increase font size
plt.rcParams['font.size'] = 14
```

### Important things for plotting.
1. Data (`df`)
2. Types of plots?
3. `x-axis`
4. `y-axis`
5. grouping (`hue`/`color`) based on categorical variables 
6. Size of the plot
7. Color palette
8. Style of the plot
9. How to interpret?

```python
# Import Data
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
# Scatter Plot in sns
# Age on x-axis ; Fare on y-axis
sns.scatterplot(x='age', y='fare', data=df)
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
# Scatter Plot in sns
# Age on x-axis ; Fare on y-axis
# Grouped by 'Sex' 
sns.scatterplot(x = 'age' , y = 'fare' , data = df , hue = 'sex')
# Do labels
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
# Change colors for male and female in scatter plot
palette = {'male' : 'black' , 'female' : 'red'}
sns.scatterplot(data = df , x = 'age' , y = 'fare' , hue = 'sex' , palette=palette)
# Do labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare vs Age')
```

## Box Plot
> For Box/Boxen Plots, we usually take Categorical Variable on X-Axis, and Numerical Value on Y-Axis. Hue is always taken to be Categorical Variable.

```python
# Change colors for male and female in scatter plot
palette = {'First' : 'black' , 'Second' : 'red' , 'Third' : 'cyan'}
sns.boxplot(data = df , x = 'sex' , y = 'age' , hue = 'class' , palette=palette)

# change legend title
plt.legend(title = 'Class' , loc = 'upper right' , prop = {'size' : 8})
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Box Plot of Age vs. Gender')
```

## Boxen Plot

```python
# Change colors for male and female in scatter plot
palette = {'First' : 'black' , 'Second' : 'red' , 'Third' : 'cyan'}
sns.boxenplot(data = df , x = 'sex' , y = 'age' , hue = 'class' , palette=palette)

# Do labels
plt.xlabel('Gender')
plt.ylabel('Age')
# change legend title
plt.legend(title = 'Class' , loc = 'upper right' , prop = {'size' : 8})
plt.title('Boxen Plot of Age vs. Gender')
```

## Bar Plot

```python
# Change colors for male and female in scatter plot
palette = {'First' : 'black' , 'Second' : 'red' , 'Third' : 'cyan'}
sns.barplot(data = df , x = 'sex' , y = 'age' , hue = 'class' , palette=palette)

# Do labels
plt.xlabel('Gender')
plt.ylabel('Age')
# change legend title
plt.legend(title = 'Class' , loc = 'upper right' , prop = {'size' : 8})
plt.title('Bar Plot of Age vs. Gender')
```
